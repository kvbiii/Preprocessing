<h1>Import libraries<h1>

```python
from Standard_Scaler import *
from Min_Max_Scaler import *
from Quantile_Transformer import *
from Robust_Scaler import *
from Plots.Preprocessing_plots import *
Plots = Preprocessing_plots()
```

$\text{Firstly, let's define data that we will scale on.}$

```python
data = np.random.randint(1, 100, size=(1000, 1))
```

<h2>Standard Scaler<h2>

$${\hat{X}}_{m,i}=\frac{X_{m,i}-{\bar{X}}_m}{\sigma_m}$$
$\text{Where: } {\hat{X}}_{m,i} \text{ - standardized observation of variable } m,$<p>
$X_{mi}  \text{ - observation } i \text{ of variable } m,$<p>
${\bar{X}}_m \text{ - arithmetic mean of variable } m,$<p>
$\sigma_m \text{ - standard deviation of variable } m.$<p>

```python
scaler = Standard_Scaler()
data_scaled = scaler.fit_transform(data)
```

```python
Plots.hist_plot_with_kde(data=data_scaled, name="Standard Scaler", bin_size=0.1)
```

$\text{Advantages of StandardScaler:}$<p>
$\text{- Simplicity of application and explainability of the scaler.}$<p>
$\text{Disadvantages of StandardScaler:}$<p> 
$\text{- Large influence of outliers,}$<p>
$\text{- Assumption of normal distribution.}$<p>

<h2>MinMaxScaler<h2>

$${\hat{X}}_{m,i}=\frac{X_{m,i}-\min{\left(X_m\right)}}{\max{\left(X_m\right)}-\min{\left(X_m\right)}}$$
$\text{Where: } {\hat{X}}_{m,i} \text{ - normalized observation } i \text{ of variable } m,$<p>
$X_{m,i} \text{ - observation } i \text{ of variable } m,$<p>
$\min{\left(X_m\right)} \text{ - minimum for variable } m,$<p>
$\max{\left(X_m\right)} \text{ - maximum for variable } m.$

```python
scaler = Min_Max_Scaler()
data_scaled = scaler.fit_transform(data)
```

```python
Plots.hist_plot_with_kde(data=data_scaled, name="MinMax Scaler", bin_size=0.1)
```

$\text{Advantages of MinMaxScaler:}$<p>
$\text{- Simplicity of application.}$<p>
$\text{Disadvantages of MinMaxScaler:}$<p> 
$\text{- Large impact of outliers,}$<p>
$\text{- Reduces the set to the interval } \left\langle0,\ 1\right\rangle \text{ by which a large amount of information is lost.}$<p>

<h2>Quantile Transformer<h2>

$\text{The process goes as follows:}$<p>
$1. \text{Specify number of quantiles.}$<p>
$2. \text{Depending on the number of quantiles we specify, the corresponding percentiles are determined, which will be a kind of rank.}$<p>
$3. \text{We choose the output distribution of the data (normal or uniform).}$<p><br>
$\text{Normal distribution:}$
$${\hat{X}}_{m,i}=F_m^{-1}\left(\mathrm{\Phi}_m\left(X_{m,i}\right)\right)$$
$\text{Where: } \mathrm{\Phi}_m \text{ - is the normal empirical distribution of the variable } m,$<p>
$F_m^{-1}\ \text{ - is the inverse normal standard distribution of the variable } m.$<p><br>
$\text{Uniform distribution:}$
$${\hat{X}}_{m,i}=\mathrm{\Phi}_m\left(X_{m,i}\right)$$

```python
scaler = Qunatile_Transformer(distribution="normal")
data_scaled = scaler.fit_transform(data)
```

```python
Plots.hist_plot_with_kde(data=data_scaled, name="Quantile Transformer (normal)", bin_size=0.1)
```

```python
scaler = Qunatile_Transformer(distribution="uniform")
data_scaled = scaler.fit_transform(data)
```

```python
Plots.hist_plot_with_kde(data=data_scaled, name="Quantile Transformer (uniform)", bin_size=0.1)
```

$\text{Advantages of Quantile Transformer:}$<p>
$\text{- Handles outliers well (doesn't put as much weight on them),}$<p>
$\text{- For a uniform distribution and a large number of observations, returns a perfectly balanced data set.}$<p>
$\text{Disadvantages of Quantile Transformer:}$<p> 
$\text{- Difficulty of interpretation (hard to predict how many quantiles should be used).}$<p>

<h2>Robust Scaler<h2>

$${\hat{X}}_{m,i}=\frac{X_{m,i}-med\left(X_m\right)}{Q_{75}{\left(X_m\right)}-Q_{25}\left(X_m\right)}$$
$\text{Where: } {X_{X}}_{m,i} \text{ - normalized observation } i \text{ for variable } m,$<p>
$X_{m,i} \text{ - observation } i \text{ of variable } m,$<p>
$med\left(X_m\right) \text{ - median for variable } m,$<p>
$Q_{75}{\left(X_m\right)} \text{ - quartile of order 3 (75th quantile) of variable } m,$<p>
$Q_{25}\left(X_m\right) \text{ - quartile of order 1 (25th quantile) of variable } m.$

```python
scaler = Robust_Scaler()
data_scaled = scaler.fit_transform(data)
```

```python
Plots.hist_plot_with_kde(data=data_scaled, name="Robust Scaler", bin_size=0.1)
```

$\text{Advantages of Robust Scaler:}$<p>
$\text{- Simplicity of application,}$<p>
$\text{- Prevents skewness and is good when there are outliers in the dataset.}$<p>
$\text{Disadvantages of Robust Scaler:}$<p> 
$\text{- The center of attention is in the parts with the most data.}$<p>
